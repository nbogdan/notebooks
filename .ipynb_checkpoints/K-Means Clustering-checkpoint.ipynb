# K-Means clustering

K-Means clustering aims to partition __n observations__ into __k clusters__ in which each observation belongs to the cluster with the nearest mean.



```python
from IPython.display import Image
Image(filename="./kmeans_alg.png")
```

The algorithm goes through several iterations until:

a.) a stable configuration is reached

b.) a timeout is reached
